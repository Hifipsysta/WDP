```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
order = np.linspace(start = 1, stop = 10, num = 100, endpoint = True)
```

```python
def LaplaceMechanismForRDP(order, scale):
    if order > 1:
        tmp1 = (1/(order-1))
        tmp2 = order/(2*order-1) * np.exp((order-1)/scale)
        tmp3 = (order - 1)/(2*order - 1) * np.exp(-order/scale)
        return tmp1 * np.log(tmp2 + tmp3)
    elif order == 1:
        tmp1 = 1/scale
        tmp2 = np.exp(-1/scale)-1
        return tmp1 + tmp2
    else:
        raise ValueError('Order cannot less than 1')
```

```python
plt.plot(order, [LaplaceMechanismForRDP(i, 5) for i in order], label = '$\lambda = 5$')
#plt.plot(order, LaplaceMechanismForWDP(order, 4), label = 'scale = 4')
plt.plot(order, [LaplaceMechanismForRDP(i, 3) for i in order], label = '$\lambda = 3$')
plt.plot(order, [LaplaceMechanismForRDP(i, 2) for i in order], label = '$\lambda = 2$')
plt.plot(order, [LaplaceMechanismForRDP(i, 1) for i in order], label = '$\lambda = 1$')
plt.xlim(left=1, right=10)
plt.ylim(bottom=0, top=2)
plt.xticks(fontsize=14.5)
plt.yticks(fontsize=14.5)
plt.xlabel(r'$\alpha$', fontsize=17)
plt.ylabel('$\epsilon$', fontsize=17)
plt.legend(fontsize=14.5, loc = 1)
```

```python
def LaplaceMechanismForWDP(order, scale):
    return (1/2) * ((np.sqrt(2 * (1/scale + np.exp(-1/scale)-1)))**(1/order))
```

```python
plt.plot(order, LaplaceMechanismForWDP(order, 5), label = '$\lambda = 5$')
#plt.plot(order, LaplaceMechanismForWDP(order, 4), label = 'scale = 4')
plt.plot(order, LaplaceMechanismForWDP(order, 3), label = '$\lambda = 3$')
plt.plot(order, LaplaceMechanismForWDP(order, 2), label = '$\lambda = 2$')
plt.plot(order, LaplaceMechanismForWDP(order, 1), label = '$\lambda = 1$')
plt.xlim(left=1, right=10)
plt.ylim(bottom=0, top=2)
plt.xticks(fontsize=14.5)
plt.yticks(fontsize=14.5)
plt.xlabel('$\mu$', fontsize=17)
plt.ylabel('$\epsilon$', fontsize=17)
plt.legend(fontsize=14.5, loc=1)
```

```python
def GaussianMechanismForRDP(order, scale):
    return order/(2*scale**2)
```

```python
plt.plot(order, GaussianMechanismForRDP(order, 5), label='$\sigma=5$')
plt.plot(order, GaussianMechanismForRDP(order, 3), label='$\sigma=3$')
plt.plot(order, GaussianMechanismForRDP(order, 2), label='$\sigma=2$')
plt.plot(order, GaussianMechanismForRDP(order, 1), label='$\sigma=1$')
plt.xlim(left=1, right=10)
plt.ylim(bottom=0, top=2)
plt.xticks(fontsize=14.5)
plt.yticks(fontsize=14.5)
plt.xlabel(r'$\alpha$', fontsize=17)
plt.ylabel(r'$\epsilon$', fontsize=17)
plt.legend(fontsize=14.5,loc=1)
```

```python
def GaussianMechanismForWDP(order, scale):
    return (1/2) * ((1/scale) ** (1/order))
```

```python
plt.plot(order, GaussianMechanismForWDP(order, 5), label='$\sigma=5$')
plt.plot(order, GaussianMechanismForWDP(order, 3), label='$\sigma=3$')
plt.plot(order, GaussianMechanismForWDP(order, 2), label='$\sigma=2$')
plt.plot(order, GaussianMechanismForWDP(order, 1), label='$\sigma=1$')
plt.xlim(left=1, right=10)
plt.ylim(bottom=0, top=2)
plt.xticks(fontsize=14.5)
plt.yticks(fontsize=14.5)
plt.xlabel('$\mu$', fontsize=17)
plt.ylabel('$\epsilon$', fontsize=17)
plt.legend(fontsize=14.5, loc=1)
```
